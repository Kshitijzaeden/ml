```python
import numpy as np
import pandas as pd
from sklearn import linear_model

import warnings
warnings.filterwarnings('ignore')
```

**DATASET FOR MAKING/TRAINING THE MODEL**

```python
df=pd.read_csv("/kaggle/input/houseprice/houseprice.csv")
print(df.head())
```

```python
df.isnull()
```

**MODEL TRAINING ** with multivariate variables

```python
model=linear_model.LinearRegression()
model.fit(df.drop('price',axis="columns"),df.price)
```

```python
model.predict([[5,5,2000,10500]])
```

```python
model.coef_
```

```python
model.intercept_
```
